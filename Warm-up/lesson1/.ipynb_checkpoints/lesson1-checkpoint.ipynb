# USE CHATGPT TO GENERATE CODE

```python
# Code a parabol in Python
import numpy as np
import matplotlib.pyplot as plt

# Define the parabola equation: y = ax^2 + bx + c
a, b, c = 1, 0, 0  # Change these values to modify the parabola
x = np.linspace(-10, 10, 400)  # X values from -10 to 10
y = a*x**2 + b*x + c  # Compute y values

# Plot the parabola
plt.plot(x, y, label=f"$y = {a}x^2 + {b}x + {c}$", color='blue')

# Formatting
plt.axhline(0, color='black', linewidth=1)  # X-axis
plt.axvline(0, color='black', linewidth=1)  # Y-axis
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.title("Parabola Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Show plot
plt.show()
```
